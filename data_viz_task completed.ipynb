```python
# Import libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
# Load data
```

```python
data = pd.read_csv("Cars93.csv")
```

### 1. Boxplot

```python
# Boxplot for revs per mile
```

```python
manufacturers = ["Audi", "Hyundai", "Suzuki", "Toyota"]
sns.boxplot(
    x = "Manufacturer",
    y = "Rev.per.mile",
    showmeans=True,  # Show mean values in the boxes
    data = data[data["Manufacturer"].isin(manufacturers)]
)
plt.xlabel("Manufacturer")
plt.ylabel("Revs per Mile")
plt.title("Box Plot of RPM by Manufacturer")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Suzuki

### 2. Histogram

```python
# Histogram of MPG
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
sns.histplot(data["MPG.city"], bins=10, label="MPG.city")
sns.histplot(data["MPG.highway"], bins=10, alpha=0.7, label="MPG.highway")  # Adjust alpha for transparency
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.title("Distribution of MPG in City and Highway")
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
its generally more cost efficient to drive on the highway

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
```

```python
plt.scatter(data["Wheelbase"], data["Turn.circle"])
plt.xlabel("Wheelbase")
plt.ylabel("Turning Circle (diameter in ft)")
plt.title("Relationship between Wheelbase and Turning Circle")
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
The relationship is a positive correlation. The Turning circle generally increases with the wheelbase sise

### 4. Barplot

```python
# Barplot for mean horsepower
```

```python
avg_hp_by_type = data.groupby("Type")["Horsepower"].mean()
plt.bar(avg_hp_by_type.index, avg_hp_by_type.values)
plt.xlabel("Car Type")
plt.ylabel("Average Horsepower")
plt.title("Average Horsepower by Car Type")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
yes
